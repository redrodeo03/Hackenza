```python
a_df = pd.read_csv('A.csv', names=['id', 'Title', 'Acronym', 'Source', 'Rank', 'Usless', 'Usless1', 'Useless2', 'Useless3'])
a_df
```

```python
a_df = a_df.drop(columns=['id', 'Source', 'Rank','Usless', 'Usless1', 'Useless2', 'Useless3'])
```

```python
a_df
```

```python
a_dict = a_df.to_dict()
a_dict = a_dict['Acronym']
a_dict
```

```python
a_list = list(a_dict.values())
print(a_list)
```

```python
a_result = result[result['booktitle'].isin(a_list)]
a_result
```

```python
pattern = '|'.join(a_list)
mask = result['booktitle'].str.contains(pattern, case=True, na=False)
# Replace NaN/NA values in the mask with False
mask = mask.fillna(False)

# Filter the DataFrame based on the mask
filtered_df = result[mask]
filtered_df
```

```python
unique = filtered_df['booktitle'].unique()
unique.sort()
print(unique)
print(len(unique))
```
